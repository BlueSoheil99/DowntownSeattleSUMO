```python
import pandas as pd
import xml.etree.ElementTree as ET
import geopandas as gpd
```

```python
# Load shapefiles
taz_data = gpd.read_file('taz2010.shp')

# load sumo tazs
tree = ET.parse('Taz_with_pseudo_link.add.xml')
root = tree.getroot()
sumo_taz = []
for taz in root.findall('taz'):
    tazid = taz.get('id')
    if tazid != 'taz_13':
        sumo_taz.append(int(tazid))
```

```python
taz_data['region'] = taz_data.COUNTY_NM
taz_data.region[taz_data.TAZ.isin(sumo_taz)] ='sumo'
```

```python
taz_data.to_file('modified shp file/taz2010.shp')
```
